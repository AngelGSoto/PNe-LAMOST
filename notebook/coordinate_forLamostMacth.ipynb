```python
from astropy.table import Table
import numpy as np
import pandas as pd
import argparse
import os
```

```python
df = pd.read_csv("../Luis_arzd099bej.csv")
```

```python
df
```

```python
# Convert the Pandas DataFrame to an Astropy Table
table = Table.from_pandas(df)
```

```python
# Mask
mask_t = table["PNstat"] == "T"
mask_p = table["PNstat"] == "P"
mask_l = table["PNstat"] == "L"
mask_pl = mask_p + mask_l

table_t = table[mask_t]
table_p = table[mask_p]
table_l = table[mask_l]
table_pl = table[mask_pl]

print("Number of true PN:", len(table_t)) 
print("Number of probable PN:", len(table_p)) 
print("Number of likely PN:", len(table_l)) 
print("Number of probable and likely PN:", len(table_pl)) 
```

```python
n_t = len(table_t["RAJ2000"])
sep_t = np.linspace(2.0, 2.0, num=n_t)
ra_t = table_t["DRAJ2000"]
dec_t = table_t["DDECJ2000"]
table_t_ = Table([ra_t, dec_t, sep_t], names=('ra', 'dec', 'radius'), meta={'name': 'first table'})
```

```python
# Save the file
table_t_.write("../True-PN/tPN-lamostMatch-2arc.dat", format="ascii.commented_header", delimiter=',', overwrite=True)
```

```python
n_pl = len(table_pl["RAJ2000"])
sep_pl = np.linspace(2.0, 2.0, num=n_pl)
ra_pl = table_pl["DRAJ2000"]
dec_pl = table_pl["DDECJ2000"]
table_pl_ = Table([ra_pl, dec_pl, sep_pl], names=('ra', 'dec', 'radius'), meta={'name': 'first table'})
```

```python
# Save the file
table_pl_.write("../Likely-Probable-PN/plPN-lamostMatch-2arc.dat", format="ascii.commented_header", delimiter=',', overwrite=True)
```
