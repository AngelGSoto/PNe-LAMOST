```python
from astropy.table import Table
import glob
```

```python
file_list = glob.glob("../better-model-final-noHalpha/*.ecsv")
```

```python
file_list 
```

```python
for file_name in file_list:
    tab = Table.read(file_name)
    mask = tab["Chi red"] < 1.59
    better_model = tab[mask]
    print(better_model)
```

```python
# Initialize variables to store the minimum value and the corresponding file
min_chi = None
min_chi_file = None

for file_name in file_list:
    tab = Table.read(file_name)

    # Check if the table is not empty
    if len(tab) > 0:
        # Make sure the column name matches the case in your data
        min_val = tab["chi"].min()

        # Check if this is the new minimum value
        if min_chi is None or min_val < min_chi:
            min_chi = min_val
            min_chi_file = file_name

# Print the file with the unique minimum "Chi" value
if min_chi_file is not None:
    print(f"File with the unique minimum 'Chi' value: {min_chi_file}")
    print(f"Minimum 'Chi' value: {min_chi}")
else:
    print("No files with 'Chi' values found.")
```
